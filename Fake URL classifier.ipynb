```python
#importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#importing the dataset
df = pd.read_csv('dataset.csv')
df
```

```python
#dropping the additional index column
df.drop('index', axis='columns', inplace=True)
df
```

```python
#some analysis on the dataset

df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# Changing the -1 in "Redirect" to 0, in order to make the 2 classes 1 (Legitimate), and 0 (Fake)

for i in range(0,len(df['Result'])):
            
    if df['Result'][i] == -1:
            
        df["Result"][i] = 0
```

```python
#checking if there is much difference in the 2 classes counts
df['Result'].value_counts()
```

```python
# plotting the counts of the Result column

plt.bar([1,0],list(df['Result'].value_counts()),color="green")
plt.title("Results Count")
plt.xlabel("classes")
plt.ylabel("count")
plt.show()
```

```python
# Printing the unique value or "classes" of each column or feature:

for col in df.columns:
    unique_list = df[col].unique()

    print('{} contains the values:\t\t{}'.format(col,unique_list))
    
# 1 => Legitimate
# 0 => suspicious
# -1 => fake or phishing
```

```python
#Features
X = df.iloc[:, : -1].values

#Target 
y = df.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
#splitting the dataset with 20% test data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 33)

print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
#importing the model evaluating functions
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
## Logistic Regression

from sklearn.linear_model import LogisticRegression
classifier1 = LogisticRegression(random_state = 1)

classifier1.fit(X_train, y_train)
```

```python
#predicting using the test data

y_pred1 = classifier1.predict(X_test)
y_pred1
```

```python
#evaluating by comparing with the test data

cm1 = confusion_matrix(y_test, y_pred1)
print("Logistic Regression Cofusion matrix:")
print(cm1)
print("\n")
print("\n")
print("Logistic Regression classification report:")
cr1 = classification_report(y_test, y_pred1)
print(cr1)
print("\n")
print("\n")
accuracy1 = accuracy_score(y_test, y_pred1)
print("Logistic Regression accuracy is {:.2%}".format(accuracy1))
```

```python
## K Nearest Neighbors (KNN)

from sklearn.neighbors import KNeighborsClassifier
classifier2 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)

classifier2.fit(X_train, y_train)
```

```python
#predicting using the test data

y_pred2 = classifier2.predict(X_test)
y_pred2
```

```python
#evaluating by comparing with the test data

cm2 = confusion_matrix(y_test, y_pred2)
print("KNN Cofusion matrix:")
print(cm2)
print("\n")
print("\n")
print("KNN classification report:")
cr2 = classification_report(y_test, y_pred2)
print(cr2)
print("\n")
print("\n")
accuracy2 = accuracy_score(y_test, y_pred2)
print("KNN accuracy is {:.2%}".format(accuracy2))
```

```python
## SVM

from sklearn.svm import SVC
classifier3 = SVC(kernel = 'linear', random_state = 0)

classifier3.fit(X_train, y_train)
```

```python
#predicting using the test data

y_pred3 = classifier3.predict(X_test)
y_pred3
```

```python
#evaluating by comparing with the test data

cm3 = confusion_matrix(y_test, y_pred3)
print("SVM Cofusion matrix:")
print(cm3)
print("\n")
print("\n")
print("SVM classification report:")
cr3 = classification_report(y_test, y_pred3)
print(cr3)
print("\n")
print("\n")
accuracy3 = accuracy_score(y_test, y_pred3)
print("SVM accuracy is {:.2%}".format(accuracy3))
```

```python
## Random Forest

from sklearn.ensemble import RandomForestClassifier
classifier4 = RandomForestClassifier(n_estimators = 1000, criterion = 'entropy', random_state = 0)

classifier4.fit(X_train, y_train)
```

```python
#predicting using the test data

y_pred4 = classifier4.predict(X_test)
y_pred4
```

```python
#evaluating by comparing with the test data

cm4 = confusion_matrix(y_test, y_pred4)
print("Random Forest Cofusion matrix:")
print(cm4)
print("\n")
print("\n")
print("Random Forest classification report:")
cr4 = classification_report(y_test, y_pred4)
print(cr4)
print("\n")
print("\n")
accuracy4 = accuracy_score(y_test, y_pred4)
print("Random Forest accuracy is {:.2%}".format(accuracy4))
```

```python
#comapring the performance of the 4 models using a bar chart

a = ['Logetic\nRegression', 'KNN', 'SVM', 'Random\nForest']
b = [accuracy1*100, accuracy2*100, accuracy3*100, accuracy4*100]

plt.figure(figsize=(10, 6))
plt.bar(a, b, color= 'blue')
plt.title('Accuracy scores of the different models')
plt.xlabel('classification model')
plt.ylabel('accuracy %')
plt.show()
```
